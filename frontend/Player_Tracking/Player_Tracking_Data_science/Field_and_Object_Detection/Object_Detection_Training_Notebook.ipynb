```python
from ultralytics import YOLO

# Load a pretrained model
model = YOLO('yolov8n.pt')  # Using YOLOv8x as per your earlier request, change to 'yolov8n.pt' if you prefer the smaller model

# Train the model
results = model.train(
    data=r'/content/football-players-detection-12/data.yaml',  # Adjusted path for your current project
    epochs=100,
    imgsz=640,
    batch=16,
    name='football_players_detection',
    patience=15,  # Early stopping patience
    save=True,    # Save best model
        # Use GPU. Set to 'cpu' if you don't have a GPU
)

# Evaluate the model
metrics = model.val()

print(metrics)

# The best model is automatically saved during training
print("Best model saved as: ", model.best)
```

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow # Import the cv2_imshow function from google.colab.patches

model = YOLO('/content/runs/detect/football_players_detection/weights/best.pt')  # path to your best weights

# Open the video file
video_path = "/content/drive/MyDrive/test vid.mp4"
cap = cv2.VideoCapture(video_path)

# Get the video's width and height
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, 30, (width, height))

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Write the frame into the output file
        out.write(annotated_frame)

        # Display the annotated frame
        cv2_imshow(annotated_frame) # Use cv2_imshow instead of cv2.imshow

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
out.release()
cv2.destroyAllWindows()
```
