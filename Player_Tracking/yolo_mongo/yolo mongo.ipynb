```python
import glob
import os
import numpy as np
import pickle

model = YOLO('yolov8n.pt')
base_path = '/content/drive/MyDrive/yolov8/test/images/'

image_paths = glob.glob(f'{base_path}/*.jpg')
print(len(image_paths))
re_dict={}
for path in image_paths:
    a=[]
    results = model.predict(source=path, conf=0.25)
    boxes = results[0].boxes.xyxy.tolist()
    classes = results[0].boxes.cls.tolist()
    names = results[0].names
    confidences = results[0].boxes.conf.tolist()
    for box, cls, conf in zip(boxes, classes, confidences):
        x1, y1, x2, y2 = box
        confidence = conf
        detected_class = cls
        name = names[int(cls)]
        a.append([name,x1, y1, x2, y2])
        re_dict[path.replace("/content/drive/MyDrive/yolov8/test/images/","")]=a

    #results[0].show()
    #results[0].save(filename="result.jpg")

print(re_dict)
with open("re_dict.pkl", "wb") as tf:
    pickle.dump(re_dict,tf)
```
