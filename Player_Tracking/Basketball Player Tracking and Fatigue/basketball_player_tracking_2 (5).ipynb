```python
# Install required packages
!pip install ultralytics deep_sort_realtime -q

# Import libraries
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
import cv2
import numpy as np
from collections import defaultdict

# Load YOLOv8 model
model = YOLO('yolov8n.pt')

# Initialize DeepSORT tracker
tracker = DeepSort(max_age=30)

# Setup video input/output
video_path = '/content/large.mp4'
output_path = 'tracked_output.mp4'
cap = cv2.VideoCapture(video_path)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Parameters and tracking data
allowed_classes = ['person']
scale_factor = 15 / 1920  # meters per pixel
player_positions = defaultdict(list)
speeds = defaultdict(list)
active_frames = defaultdict(int)
current_fatigue_scores = {}

# Frame-by-frame tracking
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, verbose=False)[0]
    detections = []

    for box in results.boxes:
        cls_id = int(box.cls[0])
        cls_name = model.names[cls_id]
        if cls_name in allowed_classes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            detections.append([[x1, y1, x2 - x1, y2 - y1], conf, cls_name])

    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        x1, y1, x2, y2 = map(int, track.to_ltrb())
        center = ((x1 + x2) / 2, (y1 + y2) / 2)

        player_positions[track_id].append(center)
        active_frames[track_id] += 1

        # Speed calculation
        if len(player_positions[track_id]) > 1:
            prev = player_positions[track_id][-2]
            dx, dy = center[0] - prev[0], center[1] - prev[1]
            dist = np.sqrt(dx**2 + dy**2)
            speed_mps = dist * fps * scale_factor
            speeds[track_id].append(speed_mps)
            cv2.putText(frame, f"Speed: {speed_mps:.2f} m/s", (x1, y2 + 20),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 2)

        color = (0, 255, 0)  # default box color

        # Fatigue analysis
        if len(speeds[track_id]) >= 25:
            recent_speeds = np.array(speeds[track_id][-25:])
            segments = np.array_split(recent_speeds, 5)
            avg_speeds = [np.mean(seg) for seg in segments]
            speed_drop = ((avg_speeds[0] - avg_speeds[-1]) / avg_speeds[0]) * 100 if avg_speeds[0] != 0 else 0
            speed_var = np.std(recent_speeds)
            duration = active_frames[track_id] / fps

            # Normalize metrics
            drop_score = min(speed_drop / 50, 1.0)
            var_score = min(speed_var / 1.5, 1.0)
            dur_score = min(duration / 20, 1.0)
            fatigue_score = 0.4 * drop_score + 0.3 * var_score + 0.3 * dur_score
            current_fatigue_scores[track_id] = fatigue_score

            # Fatigue level
            if fatigue_score < 0.3:
                fatigue_level = "Low"
                color = (0, 255, 0)
            elif fatigue_score < 0.6:
                fatigue_level = "Medium"
                color = (0, 255, 255)
            else:
                fatigue_level = "High"
                color = (0, 0, 255)

            # Fatigue label
            cv2.putText(frame, f"Fatigue: {fatigue_level}", (x1, y2 + 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

            # Warning for high fatigue
            if fatigue_score >= 0.6:
                cv2.putText(frame, "FATIGUE SPIKE", (x1, y1 - 30),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

            # Fatigue bar (horizontal)
            bar_x = x1
            bar_y = y2 + 55
            bar_width = 100
            filled = int(fatigue_score * bar_width)
            cv2.rectangle(frame, (bar_x, bar_y), (bar_x + bar_width, bar_y + 10), (200, 200, 200), 1)
            cv2.rectangle(frame, (bar_x, bar_y), (bar_x + filled, bar_y + 10), color, -1)
            cv2.putText(frame, f"{int(fatigue_score * 100)}%", (bar_x + bar_width + 5, bar_y + 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)

        # Draw bounding box and ID
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, f"ID {track_id}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

    # Top 3 fatigued players dashboard
    sorted_scores = sorted(current_fatigue_scores.items(), key=lambda x: x[1], reverse=True)[:3]
    for i, (pid, score) in enumerate(sorted_scores):
        text = f"ID {pid} | Fatigue: {score:.2f}"
        cv2.putText(frame, text, (20, 40 + i * 20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

    # Transparent bottom overlay panel
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, height - 40), (width, height), (0, 0, 0), -1)
    alpha = 0.5
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

    frame_count = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
    avg_fatigue = np.mean(list(current_fatigue_scores.values())) if current_fatigue_scores else 0
    cv2.putText(frame, f"Frame: {frame_count} | Players: {len(current_fatigue_scores)} | Avg Fatigue: {avg_fatigue:.2f}",
                (10, height - 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    out.write(frame)

# Finalize
cap.release()
out.release()
print(f"Tracking complete. Output saved to: {output_path}")
```
