```python
#Importing Librariesfrom ultralytics import YOLO
from ultralytics.solutions import object_counter
import cv2
from IPython.display import display, Image
import numpy as np

model = YOLO("yolov8n.pt")

#Reading the Videocap = cv2.VideoCapture("/content/peoples.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

#Assigning the points for Region of Interestregion_points = [(20, 400), (1080, 404), (1080, 360), (20, 360)]

counter = object_counter.ObjectCounter()
counter.set_args(view_img=True,
                 reg_pts=region_points,
                 classes_names=model.names,
                 draw_tracks=True)

processed_frames = []

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    tracks = model.track(im0, persist=True, show=False)
    im0 = counter.start_counting(im0, tracks)
    im0_rgb = cv2.cvtColor(im0, cv2.COLOR_BGR2RGB)
    processed_frames.append(im0_rgb)

cap.release()

video_writer = cv2.VideoWriter("object_counting_output.mp4",
                               cv2.VideoWriter_fourcc(*'mp4v'),
                               fps,
                               (w, h))

for frame in processed_frames:
    video_writer.write(frame)

video_writer.release()
cv2.destroyAllWindows()
```
