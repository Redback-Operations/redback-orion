```python
import pandas as pd
import matplotlib as plt
# Read the CSV file
df = pd.read_csv('Heart_rate.csv')

# # Filter out rows where heart rate values do not start with "Heart Rate Value"
cleaned_df = df[df['data values'].str.startswith('Heart Rate Value')]

cleaned_df.loc[:, 'data values'] = cleaned_df['data values'].str.replace('Heart Rate Value (BPM) = ', '').astype(int)

cleaned_df.to_csv('Cleaned_Heart_rate.csv', index=False)
```

```python
df = pd.read_csv('Cleaned_Heart_rate.csv')

# Rename the column
df.rename(columns={'data values': 'HeartRate'}, inplace=True)
df['time'] = pd.to_datetime(df['time'])
# Write the modified DataFrame back to a CSV file
df.to_csv('Cleaned_Heart_rate.csv', index=False)
```

```python
## Visualisartion
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('Cleaned_Heart_rate.csv')

# Assuming 'time' is the index and 'HeartRate' is the column you want to plot
# If 'time' is not the index, you can set it as index using df.set_index('time', inplace=True)
# Make sure 'time' column is in datetime form
df['time'] = pd.to_datetime(df['time'])
# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(df['time'], df['HeartRate'], marker='o', linestyle='-')  # Line plot with markers
plt.title('Heart Rate over Time')
plt.xlabel('Time')
plt.ylabel('Heart Rate')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
